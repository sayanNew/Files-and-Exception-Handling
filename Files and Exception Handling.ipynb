```python
#1. What is the difference between interpreted and compiled languages?

'''
1) In compiled Languages, the entire source code is translated into machine code before execution using a compiler. In interpreted languages, code is translated and executed line by line at runtime using an interpreter.
2) In compiled Languages, execution is generally faster because the code is already in machine language. In interpreted languages, execution is usually slower compared to compiled languages.
3) In compiled Languages, errors are usually reported after compilation. In interpreted languages, errors are reported as soon as the faulty line is reached.
'''
```

```python
#6.  Write a Python program that logs an error message to a log file when a division by zero exception occurs.

import logging

# Configure logging
logging.basicConfig(
    filename="error.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        logging.error("Division by zero error occurred while dividing %s by %s", a, b)
        return None

# Example usage
result = divide(10, 0)
print("Result:", result)
```

```python
#14. Write a program that uses the logging module to log both informational and error messages

import logging

# Configure logging
logging.basicConfig(
    filename="app.log",          # Log file name
    level=logging.INFO,           # Minimum level to log (INFO and above)
    format="%(asctime)s - %(levelname)s - %(message)s"
)

# Log an informational message
logging.info("Application started successfully.")

try:
    # Example operation that may fail
    x = 10
    y = 0
    result = x / y  # This will raise ZeroDivisionError
    logging.info(f"Division result: {result}")

except ZeroDivisionError:
    # Log an error message
    logging.error("Division by zero occurred.", exc_info=True)

# Another informational message
logging.info("Application finished execution.")
```

```python
#16. Demonstrate how to use memory profiling to check the memory usage of a small program.

import tracemalloc
import time

def create_large_list():
    a = [i for i in range(10_000_000)]
    time.sleep(1)
    return a

# Start tracking memory
tracemalloc.start()

# Run your function
create_large_list()

# Take a snapshot
current, peak = tracemalloc.get_traced_memory()

print(f"Current memory usage: {current / 1024 / 1024:.2f} MB")
print(f"Peak memory usage: {peak / 1024 / 1024:.2f} MB")

# Stop tracking
tracemalloc.stop()
```

```python
#18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?

import logging
from logging.handlers import RotatingFileHandler

# Clear existing handlers (important for Jupyter to avoid duplicate logs)
logger = logging.getLogger("demo")
logger.handlers.clear()

logger.setLevel(logging.INFO)

handler = RotatingFileHandler(
    "demo.log",
    maxBytes=1024 * 1024,   # 1 MB
    backupCount=2
)

formatter = logging.Formatter(
    "%(asctime)s - %(levelname)s - %(message)s"
)

handler.setFormatter(formatter)
logger.addHandler(handler)

# Generate logs to trigger rotation
for i in range(20000):
    logger.info(f"This is log message number {i}")

print("Logging completed successfully!")
```
